```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# 1. Collecte des Données
data = yf.download('AAPL', start='2010-01-01', end='2023-10-10')
prices = data['Close'].values.reshape(-1, 1)

# 2. Prétraitement des Données
scaler = MinMaxScaler(feature_range=(0, 1))
prices_scaled = scaler.fit_transform(prices)

# 3. Création des Séquences Temporelles
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

seq_length = 60
X, y = create_sequences(prices_scaled, seq_length)

# 4. Division des Données
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# 5. Construction du Modèle LSTM
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(seq_length, 1)))
model.add(Dropout(0.2))
model.add(LSTM(50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(25))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.summary()

# 6. Entraînement du Modèle
history = model.fit(X_train, y_train, batch_size=32, epochs=20, validation_data=(X_test, y_test))

# 7. Évaluation du Modèle
predictions = model.predict(X_test)
predictions = scaler.inverse_transform(predictions)
y_test_actual = scaler.inverse_transform(y_test)

# Calcul des métriques
rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
mae = mean_absolute_error(y_test_actual, predictions)
r2 = r2_score(y_test_actual, predictions)

print(f"RMSE: {rmse:.2f}")
print(f"MAE: {mae:.2f}")
print(f"R²: {r2:.2f}")

# Visualiser les résultats
plt.figure(figsize=(12, 6))
plt.plot(y_test_actual, label='Prix Réel', color='blue')
plt.plot(predictions, label='Prix Prédit', color='red', linestyle='--')
plt.title('Prédictions vs Prix Réels (Test Set)')
plt.xlabel('Jours')
plt.ylabel('Prix de Clôture')
plt.legend()
plt.show()

# 8. Backtesting
# Simuler des trades basés sur les prédictions
returns = np.diff(y_test_actual.flatten()) / y_test_actual[:-1]
strategy_returns = np.where(predictions[:-1] > y_test_actual[:-1], returns, -returns)

# Calculer le rendement cumulatif
cumulative_returns = np.cumprod(1 + strategy_returns) - 1

# Visualiser les rendements
plt.figure(figsize=(12, 6))
plt.plot(cumulative_returns, label='Stratégie LSTM', color='green')
plt.title('Rendement Cumulatif de la Stratégie LSTM')
plt.xlabel('Jours')
plt.ylabel('Rendement Cumulatif')
plt.legend()
plt.show()

# 9. Prévisions pour les 20 jours d'octobre 2023 (du 11-10-2023 au 30-10-2023)
# Préparer la séquence pour la prédiction
last_sequence = prices_scaled[-seq_length:]  # Les 60 derniers jours disponibles
last_sequence = last_sequence.reshape((1, seq_length, 1))  # Reshape pour LSTM

# Liste pour stocker les prédictions
predictions_future = []

# Faire des prévisions pour les 20 jours
for _ in range(20):
    # Prédire le jour suivant
    next_day_prediction = model.predict(last_sequence)
    
    # Ajouter la prédiction à la liste
    predictions_future.append(next_day_prediction[0][0])
    
    # Redimensionner next_day_prediction pour qu'il ait la forme (1, 1, 1)
    next_day_prediction_reshaped = next_day_prediction.reshape((1, 1, 1))
    
    # Mettre à jour la séquence d'entrée
    last_sequence = np.concatenate((last_sequence[:, 1:, :], next_day_prediction_reshaped), axis=1)

# Dénormaliser les prédictions
predictions_future = scaler.inverse_transform(np.array(predictions_future).reshape(-1, 1))

# Afficher les prédictions
print("Prédictions pour les 20 jours (du 11-10-2023 au 30-10-2023) :")
for i, pred in enumerate(predictions_future):
    print(f"Jour {i+1} : {pred[0]:.2f}")

# 10. Visualiser les Prédictions Futures
# Dates pour les prédictions (du 11-10-2023 au 30-10-2023)
prediction_dates = pd.date_range(start='2023-10-11', periods=20)

# Données historiques jusqu'au 10-10-2023
historical_dates = data.index
historical_prices = data['Close']

# Visualiser les données historiques et les prédictions
plt.figure(figsize=(12, 6))
plt.plot(historical_dates, historical_prices, label='Données Historiques', color='blue')
plt.plot(prediction_dates, predictions_future, label='Prédictions', color='red', linestyle='--', marker='o')
plt.title('Prédictions du Cours Boursier AAPL pour Octobre 2023')
plt.xlabel('Date')
plt.ylabel('Prix de Clôture')
plt.legend()
plt.grid()
plt.show()
```
